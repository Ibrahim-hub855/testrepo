```python
import matplotlib.pyplot as plt
import pandas as pd
car_sales =pd.read_csv("data/car-sales.csv")
car_sales
```

```python
plt.plot("Price")
```

```python
car_sales["Price"]= car_sales["Price"].str.replace('[\$\,\.]','')
car_sales
```

```python
car_sales["Price"]= car_sales["Price"].str[:-2]
car_sales
```

```python
car_sales["Sales_Date"] = pd.date_range("1/1/2020", periods=len(car_sales))
car_sales["Total_Sales"] = car_sales["Price"].astype(int).cumsum()
car_sales
```

```python
car_sales.plot(x='Sales_Date',y='Total_Sales')
```

```python
#car_sales.plot(x= 'Odometer (KM)', y='Price', kind ='scatter')

car_sales.plot.scatter(x= 'Odometer (KM)', y='Price')
```

```python
import pandas as pd
heart_disease=pd.read_csv("data/heart-disease.csv")
heart_disease
```

```python
heart_disease.head(10)
```

```python
heart_disease["age"].plot.hist(bins=50)
```

```python
heart_disease.plot.hist(figsize=(10,40), subplots =True)
```

```python
heart_disease.plot.hist(figsize=(10,30), subplots =True)
```

```python
over_fifty = heart_disease[heart_disease["age"] >50]
over_fifty
```

```python
over_fifty.plot.scatter(x="age", y="chol", c="target", figsize=(10,6))
```

```python
fig, ax =plt.subplots(figsize=(10,6))
plt.style.use('seaborn-talk')#this is c map
#Plot the Data
scatter = ax.scatter(over_fifty["age"],over_fifty["chol"], c=over_fifty["target"])

#customize the plot
ax.set(title ="Heart disease and cholestrol level", xlabel ="Age", ylabel="Cholesterol")

ax.legend(*scatter.legend_elements(), title = "Target")

ax.axhline (over_fifty["chol"].mean(), linestyle = "--")

fig.suptitle("Government hospital Warwick, England", fontsize=16, fontweight ="bold")
fig.savefig("myplot")
```
